```python
#| echo: false
#| warning: false
#| fig-cap: Chamber pressure and thrust vs. time using web vs. surface area table

import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import interp1d

# ================================================================
# Propellant and motor parameters
# ================================================================
rho_p   = 1700.0      # kg/m³
a       = 0.0006      # burning rate coefficient (m/s / Pa^n)
n       = 0.35
c_star  = 1550.0      # m/s
A_t     = 0.0015       # m²
V_c0    = 0.0015      # m³
dt      = 0.001
t_max   = 5.0

# Grain burnback table: web (m) vs. burning surface area (m²)
web_table = np.array([0.000, 0.005, 0.010, 0.015, 0.020, 0.025, 0.030])
Ab_table  = np.array([0.050, 0.052, 0.055, 0.060, 0.058, 0.045, 0.020])

Ab_interp = interp1d(web_table, Ab_table, kind='linear', bounds_error=False, fill_value=0.0)

# Simulation code...
t_list = []
P_list = []
F_list = []
w = 0.0
V_c = V_c0
P = 0.1e6
t = 0.0

while t < t_max and w < web_table[-1]:
    A_b = Ab_interp(w)
    r = a * P**n if P > 0 else 0.0
    m_gen = rho_p * A_b * r
    m_out = P * A_t / c_star if P > 0 else 0.0
    dPdt = (c_star * 300 / V_c) * (m_gen - m_out)
    P = max(P + dPdt * dt, 0.0)
    w += r * dt
    t += dt
    
    C_F_approx = 1.55
    F = P * A_t * C_F_approx
    
    t_list.append(t)
    P_list.append(P / 1e6)
    F_list.append(F / 1000)

# Plot
fig, ax1 = plt.subplots(figsize=(10, 6))
ax1.plot(t_list, P_list, color='tab:blue', label='Chamber Pressure')
ax1.set_xlabel('Time (s)')
ax1.set_ylabel('Chamber Pressure (MPa)', color='tab:blue')
ax1.grid(True, alpha=0.3)

ax2 = ax1.twinx()
ax2.plot(t_list, F_list, color='tab:red', label='Thrust')
ax2.set_ylabel('Thrust (kN)', color='tab:red')

plt.title('0D Ballistic Prediction with Web vs. Surface Area Table')
plt.tight_layout()
plt.show()

print(f"Burn time: {t:.3f} s")
print(f"Peak pressure: {max(P_list):.2f} MPa")
print(f"Peak thrust: {max(F_list):.1f} kN")
```
